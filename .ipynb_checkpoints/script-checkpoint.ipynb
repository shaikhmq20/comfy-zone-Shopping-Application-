```python
import pandas as pd
```

```python
df = pd.read_json("products.json")
df
```

```python
prices_list = list(df["price"])
```

```python
from random import choice
```

```python
new_prices_list = []
for price in prices_list:    
    price_list = [price]
    for i in range(5):
        random_discount = choice(range(0, price // 5))
        price_list.append(price - random_discount)
        
    new_prices_list.append(price_list)

new_price_list
```
